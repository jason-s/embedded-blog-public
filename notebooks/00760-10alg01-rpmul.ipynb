```python
def rpmul(a,b):
    result = 0
    while b != 0:
        if b & 1:
            result += a
        b >>= 1
        a <<= 1
    return result

rpmul(13,27)
```

```python
def rpexp(a,b):
    result = 1
    while b != 0:
        if b & 1:
            result *= a
        b >>= 1
        a *= a
    return result

print rpexp(3,13)
print 3 ** 13
```

```python
def rpmul_gf2(a,b):
    result = 0
    while b != 0:
        if b & 1:
            result ^= a
        b >>= 1
        a <<= 1
    return result
```
